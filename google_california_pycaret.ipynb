# Быстрая модель с pycaret

Для быстроты воспользуемся AutoML с помощью pycaret

```python
!pip install pycaret
```

Не забудем и красивые графики, конечно же

```python
from pycaret.utils import enable_colab
enable_colab()

# Это активирует красивые графики в colab
```

pycaret загрузили, теперь к главному. Грузим всё для регрессии

```python
from pycaret.regression import *
```

В целом мы уже готовы. Создаём pipeline, который будет автоматически подготавливать данные к моделям. 

## Модель с пайплан по умолчанию

```python
# Берём всё по умолчанию для первоначальной оценки
reg_set = setup(data = df_train, target = 'median_house_value', silent = True, session_id=42)
```

Все подготовки сделаны, теперь обучим модели и выберем самыe хорошенькиe

Для главной характеристики выберем MAPE, чтобы смотреть ошибку в процентном соотношении

```python
raw_top5 = compare_models(sort='mape', round=2, n_select=5)
```

Лучшие модели найдены, но нет предела совершенству. Попробуем затюнить модели, а также попробуем ансамбли

```python
bagged_rf = ensemble_model(raw_top5[0], method = 'Bagging', optimize='mape')
```

```python
tune_lgbm = tune_model(raw_top5[1], round = 2, optimize='mape')
```

```python
blender = blend_models(raw_top5[1:4], round = 2, optimize='mape')
```

```python
stacker = stack_models(raw_top5[1:4], round = 2, optimize='mape')
```

Посмотрим как данные будут вести себя на тестовой выборке наши изначальные модели и улучшенные

```python
impr_models = {'bagged': bagged_rf, 'lgbm(tuned)': tune_lgbm, 'stacker':stacker, 'blender':blender}
raw_models = dict(zip(['Random Forest Regressor', 'Light Gradient Boosting Machine', 'Extra Trees Regressor', 'Gradient Boosting Regressor', 'Decision Tree Regressor'], raw_top5))
```

Список моделей

```python
from pycaret.utils import check_metric
```

```python
print('Изначальные модели:')
raw = {}
for k,v in raw_models.items():
  mape = check_metric(predict_model(v, df_test).median_house_value, predict_model(v, df_test).Label, 'MAPE')
  raw[k] = mape
raw_df = pd.DataFrame.from_dict(raw, orient='index', columns=['MAPE'])
display(rawdf.sort_values('MAPE'))

print()
print('Улучшенные модели:')
impr = {}
for k,v in impr_models.items():
  mape = check_metric(predict_model(v, df_test).median_house_value, predict_model(v, df_test).Label, 'MAPE')
  impr[k] = mape
impr_df = pd.DataFrame.from_dict(impr, orient='index', columns=['MAPE'])
display(impr_df.sort_values('MAPE'))
```

Итого ансамблевая модель, полученная с помощью стакинга стала лучшей ошибаемся в среднем на 18,11%). Сама по себе модель не полезна, но pycaret позволяет сохранить её вместе с обрабатывающим pipeline

```python
save_model(stacker, 'california_model_stacker_standard_pipeline')
```
